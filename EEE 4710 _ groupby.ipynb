```python
import pandas as pd
import numpy as np

train_df = pd.read_csv('titanic_train.csv')
```

```python
train_df
```

```python
train_df = train_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
train_df
```

**If you want to know the average age of the passengers - you can do the following**

```python
train_df['Age'].mean()
```

**Now say, you want to know - what is the average age of the female passengers as well as the male passengers**

To obtain that, you can separate the data based on gender using boolean masking - then do the average.
Or, you can use the groupby function in pandas to do it directly.

```python
# this will give u the two desired columns
train_df[['Sex','Age']]
```

```python
# this will return a groupby object. you have to add an 'aggregating function' after that.
train_df[['Sex','Age']].groupby('Age')
```

```python
train_df[['Sex','Age']].groupby('Sex').mean()
```

```python
# you can see female passengers had a higher average fair
train_df[['Sex','Fare']].groupby('Sex').mean()
```

```python
train_df.groupby('Sex').mean()
```

**Q -> what is the average age of people who survived and did not survive?**

```python
train_df[["Survived", "Age"]].groupby(['Survived']).mean()
```
